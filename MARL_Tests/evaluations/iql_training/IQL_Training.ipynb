```python
from mpe2 import simple_spread_v3
from pettingzoo.utils import aec_to_parallel
import supersuit as ss
from stable_baselines3 import DQN
import gymnasium
import os

env = simple_spread_v3.env(N=3, max_cycles=1000)
parallel_env = aec_to_parallel(env)
vec_env = ss.pettingzoo_env_to_vec_env_v1(parallel_env)
vec_env = ss.concat_vec_envs_v1(vec_env, 3, num_cpus=1, base_class='stable_baselines3')



save_path = ".iql_training/simple_spread_v3_model_{}.zip"
os.makedirs("./iql_training", exist_ok=True)
```

```python
for i in range(3):
   print(f"\n Training agent {i}...")
   model = DQN("MlpPolicy", vec_env, verbose=1, tensorboard_log="./iql_logs/")
   model.learn(total_timesteps=50_000)
   model.save(save_path.format(i))
```
